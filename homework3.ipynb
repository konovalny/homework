```python
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt('http://www.scipy-lectures.org/_downloads/populations.txt') 
year, hare, lynx, carrot = data[:, 0], data[:, 1], data[:, 2], data[:, 3]
```

# â„–1 Plot the populations for the years in the period. Add legend and axis labels to the plot.

```python
plt.figure(figsize=(16, 8))
plt.plot(year, hare, label='hare')
plt.plot(year, lynx, label='lynx')
plt.plot(year, carrot, label='carrot')
plt.xlabel('year', fontsize=14)
plt.ylabel('population', fontsize=14)
plt.xticks(year, fontsize=10)
plt.legend()
plt.show()
```

# â„–2 Find which years any of the populations is above 50000. Show time periods where populations are above 50000 in the plot of populations.

```python
plt.figure(figsize=(16, 8))
plt.plot(year, hare, label='hare')
plt.plot(year, lynx, label='lynx')
plt.plot(year, carrot, label='carrot')
plt.xlabel('year', fontsize=14)
plt.ylabel('population', fontsize=14)
plt.xticks(year, fontsize=10)
plt.legend()

plt.scatter(year[hare > 50000], hare[hare > 50000], color='blue', s=70)
plt.scatter(year[lynx > 50000], lynx[lynx > 50000], color='blue', s=70)
plt.scatter(year[carrot > 50000], carrot[carrot > 50000], color='blue', s=70)
plt.show()
print(sorted(np.hstack((year[hare > 50000], year[lynx > 50000], year[carrot > 50000]))))
```

# â„–3 Find which year each species had the largest population. Mark these years in the plot of populations.

```python
plt.figure(figsize=(16, 8))
plt.plot(year, hare, label='hare')
plt.plot(year, lynx, label='lynx')
plt.plot(year, carrot, label='carrot')
plt.xlabel('year', fontsize=14)
plt.ylabel('population', fontsize=14)
plt.xticks(year, fontsize=10)
plt.legend()

plt.scatter(year[np.argmax(hare)], hare[np.argmax(hare)], color='red', s=30)
plt.scatter(year[np.argmax(lynx)], lynx[np.argmax(lynx)], color='red', s=30)
plt.scatter(year[np.argmax(carrot)], carrot[np.argmax(carrot)], color='red', s=30)
plt.show()
print('hare:', year[np.argmax(hare)])
print('lynx:', year[np.argmax(lynx)])
print('carrot:', year[np.argmax(carrot)])
```

# â„–4 Assuming total population of hares and lynxes is 100%, plot stacked bar graph showing ratio of each specis for the years in the period.

```python
plt.figure(figsize=(16, 8))
plt.bar(year, hare/(hare+lynx)*100, label = 'hare')
plt.bar(year, lynx/(hare+lynx)*100, bottom = hare/(hare+lynx)*100, label = 'lynx')
plt.xlabel('year', fontsize=14)
plt.ylabel('population', fontsize=14)
plt.xticks(year, fontsize=12)
plt.legend()
plt.show()
```

# â„–5 Find the mean and std of the populations of each species. Plot the historgam of population for each species and show mean values with vertical line. Arrange 3 subplots in a row.

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (16, 8))
fig.tight_layout(pad = 4)

ax1.hist(hare, bins = 15, color = 'blue')
ax1.vlines(np.mean(hare), 0, 4)

ax2.hist(lynx, bins = 15, color = 'red')
ax2.vlines(np.mean(lynx), 0, 5)

ax3.hist(carrot, bins = 15, color = 'green')
ax3.vlines(np.mean(carrot), 0, 2)

fig.show()
print(np.std(hare), np.std(lynx), np.std(carrot))
```

# â„–6 Find which species (hares or lynxes) has the largest population for each year. Print the result as [H, H, L, H, ...]. Plot a pie chart showing the ratio of "H" and "L" values.

```python
highest = np.array([None]*21)
vs = hare >= lynx
highest[vs] = 'H'
highest[np.logical_not(vs)] = 'L'

size = [np.count_nonzero(hare >= lynx), 21 - np.count_nonzero(hare >= lynx)]
labels = ['hare', 'lynx']

plt.figure(figsize=(16, 8))
plt.pie(size, labels=labels)
plt.legend()
plt.show()

print(highest)
```

# â„–7 Plot the change in the hare population and the number of lynxes. Find the correlation coefficient.

```python
print(np.corrcoef(hare, lynx)[0][1])

plt.figure(figsize = (16,8))

plt.plot(year, lynx, label='lynx', zorder = -1)
plt.plot(year[:20] + 1, (hare - np.roll(hare,1))[:20], label='hare', zorder = -1)

plt.xlabel('year', fontsize=14)
plt.xticks(year)
plt.ylabel('population', fontsize=14)
plt.legend()
plt.show()
```

# â„–8 Using a scatter plot, show the population of hares vs carrots and hares vs lynxes by year on two separate subplots (each point corresponds to a particular year, and the point coordinates correspond to the population of the two species in that year).

```python
fig, axes = plt.subplots(nrows=1,ncols=2, figsize = (16,8))


axes[0].scatter(hare,carrot, label='hare vs carrot')
axes[0].set_xlabel('hare', fontsize=16)
axes[0].set_ylabel('carrot', fontsize=16)
axes[0].legend()

axes[1].scatter(hare,lynx, label='hare vs lynx')
axes[1].set_xlabel('hare', fontsize=16)
axes[1].set_ylabel('lynx', fontsize=16)
axes[1].legend()

fig. tight_layout ()
```
